### 2. Compute the Fundamental Matrix F using the 8 point algorithm

```python
#      1. Shift and Scale the Pixels
# Find the dimensions of the original image, shift them to the center and normalise
# Construct a Transformation Matrix 


#      1. Shift and Scale the Pixels
# Find the dimensions of the original image, shift them to the center and normalise
# Construct a Transformation Matrix using Hartley's normalisation method
def get_augmented(ptsL, ptsR):
    xl = np.transpose(ptsL)[0]
    yl = np.transpose(ptsL)[1]

    xl_norm = xl - np.mean(xl)
    yl_norm = yl - np.mean(yl)

    s_l = np.sqrt(1/(2*len(ptsL)) * np.sum(xl_norm**2 + yl_norm**2))

    Sl = np.diag([1/s_l,1/s_l,1])
    tl = [[1, 0, -np.mean(xl)], [0, 1, -np.mean(yl)], [0, 0, 1]]

    TL = np.array(np.dot(Sl, tl)) # Transformation Matrix


    xr = np.transpose(ptsR)[0]
    yr = np.transpose(ptsR)[1]

    xr_norm = xr - np.mean(xr)
    yr_norm = yr - np.mean(yr)

    s_r = np.sqrt(1/(2*len(ptsL)) * np.sum(xr_norm**2 + yr_norm**2))

    Sr = np.diag([1/s_r,1/s_r,1])
    tr = [[1, 0, -np.mean(xr)], [0, 1, -np.mean(yr)], [0, 0, 1]]

    TR = np.array(np.dot(Sr, tr)) # Transformation Matrix

    # For every set of pixels in the matched features 
    ptsL_augment = np.array([np.dot(TL, np.transpose([x,y,1]))[0:2] for x,y in ptsL])
    ptsR_augment = np.array([np.dot(TR, np.transpose([x,y,1]))[0:2] for x,y in ptsR])
    
    return ptsL_augment, ptsR_augment, TL, TR

ptsL_augment, ptsR_augment, TL, TR = get_augmented(ptsL, ptsR)

```

```python
#      2. Select 8 random points and construct design matrix
# Create a mask to select 8 random pixels

def find_F(ptsL_augment, ptsR_augment, TL, TR, mask=True, svd=True):
    success = True
    if mask:
        index=np.random.choice(ptsL_augment.shape[0], 8, replace=False)

        left_points = ptsL_augment[index] # 8 Matched points from left image
        right_points = ptsR_augment[index] # 8 matched points from right image

    
    else:
        # Assume pts are all inliers
        left_points = ptsL_augment # Use Matched points from left image
        right_points = ptsR_augment # Use matched points from right image
    # construct design matrix
    design = []

    for i in range(0, left_points.shape[0]):
        p1,p2 = left_points[i]
        q1,q2 = right_points[i]
        design.append([p1*q1, p2*q1, q1, p1*q2, p2*q2, q2, p1, p2, 1])
        
    
    if np.linalg.matrix_rank(design) < 8:
        print('Rank not satisfied')
        print(np.linalg.matrix_rank(design) )

        
    #       3. Perform SVD of design matrix 
    U,s,Vh = np.linalg.svd(design) # SVD decomposition
    F_vector = Vh[-1] # Right Nullspace F_vector is the smallest row of design matrix

    #.      4. Compose the draft matrix

    draft = np.array([F_vector[0:3], F_vector[3:6], F_vector[6:9]])


    #       5. Make draft singular by taking SVD and reassembling
    Ud,sd,Vhd = np.linalg.svd(draft) # SVD decomposition

    # Force smallest value of s to be 0
    if svd:
        sd[-1] = 0
    diag_sd = np.diag(sd)

    # Reconstruct draft F
    F_hat = np.dot(Ud, np.dot(diag_sd, Vhd)) 

    # Translate F_hat back to image plane
    F = np.dot(np.transpose(TR), np.dot(F_hat,TL))
    
    # Calculate number of inliers
    ptsL_filt = []
    ptsR_filt = []
    
    if len(ptsL) > 100:
        # There are quite a few matched features
        d_threshold = 1
    elif len(ptsL) > 45:
        # Not many matched features
        d_threshold = 2
    else:
        # Very few matched features
        d_threshold = 3
    
    count = 0
    for i in range(0, len(ptsL)):
        
        p1,p2 = ptsL[i]
        q1,q2 = ptsR[i]

        line_R = np.dot(F,np.transpose([p1,p2,1])) # Use left point
        
        dr = np.abs(line_R[0]*q1 + line_R[1]*q2 + line_R[2])/np.sqrt(line_R[0]**2 + line_R[1]**2 )

        if dr <= d_threshold:
            count += 1
            ptsL_filt.append(ptsL[i])
            ptsR_filt.append(ptsR[i])

    ptsL_filt = np.array(ptsL_filt)
    ptsR_filt = np.array(ptsR_filt)
    
    return F, count, ptsL_filt, ptsR_filt

# Finds the best Fundamental Matrix using RANSAC
def find_best_F(ptsL_augment, ptsR_augment, TL, TR):

    # Create an initial estimate that gets compared to
    F_best, best_count, ptsL_filt_best, ptsR_filt_best = find_F(ptsL_augment, ptsR_augment, TL, TR)

    # Estimate number of trials required 
    p = 0.99

    e = 0.5 # assume 60% of points are outliers
    s = 8 # Number of points required 
    num_trials = int(np.log(1-p)/(np.log(1-(1-e)**s))) # Estimate number of trials taken
    print(f'There are {num_trials} being used for RANSAC')

    # Perform RANSAC to find cases. If not success then 
    for i in range(0,num_trials):
        F, count, ptsL_filt, ptsR_filt  = find_F(ptsL_augment, ptsR_augment, TL, TR)

        if count > best_count:
            F_best = F 
            best_count = count
            ptsL_filt_best = ptsL_filt
            ptsR_filt_best = ptsR_filt

    # Transform best inliers back to normalised coordinates for the calculation of F
    ptsL_filt_best_transformed = np.array([np.dot(TL, [x, y, 1])[0:2] for x,y in ptsL_filt_best])
    ptsR_filt_best_transformed = np.array([np.dot(TR, [x, y, 1])[0:2] for x,y in ptsR_filt_best])

    print(f'{len(ptsR_filt_best_transformed)} inliers used to recompute F')

    # Use best transformed inliers to recompute F 
    F_best, best_count, ptsL_filt_best, ptsR_filt_best = find_F(ptsL_filt_best_transformed, ptsR_filt_best_transformed, TL, TR, False)
        
    return F_best, ptsL_filt_best, ptsR_filt_best

F_best, ptsL_filt_best, ptsR_filt_best = find_best_F(ptsL_augment, ptsR_augment, TL, TR)

print(F_best)
print(f'F has determinant = {np.linalg.det(F_best)}')

print(f'There are {len(ptsL_filt_best)} inliers / {len(ptsL)}')
```
